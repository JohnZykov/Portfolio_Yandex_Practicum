# Проект «Прогнозирование оттока клиентов оператора связи»

# Описание проекта

Оператор связи «Ниединогоразрыва.ком» хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах.

Цель проекта — проанализировать данные, подготовить их для последующего использования и создать модель для прогнозирования оттока клиентов. 

Оператор предоставляет два основных типа услуг: 

1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL*,* от англ. *digital subscriber line*, «цифровая абонентская линия») или оптоволоконный кабель (*Fiber optic*).  

Также доступны такие услуги:

- Интернет-безопасность: антивирус (*DeviceProtection*) и блокировка небезопасных сайтов (*OnlineSecurity*);
- Выделенная линия технической поддержки (*TechSupport*);
- Облачное хранилище файлов для резервного копирования данных (*OnlineBackup*);
- Стриминговое телевидение (*StreamingTV*) и каталог фильмов (*StreamingMovies*).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

```python
#Быстрое построение барчарта для категориальных переменных
def fast_barchart(column, t='Заголовок', x='Значение', y='Количество'):
    plt.figure(figsize=(6,3))
    sns.countplot(data=merged_data, x=column)
    plt.title(t)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.xticks(rotation=45)
    plt.show()
    

#Быстрое построение гистограммы для числовых переменных 
def fast_hist(column, t='Заголовок', x='Значение', y='Количество'):
    plt.figure(figsize=(8, 4))
    sns.histplot(data=merged_data[merged_data['Target'] == 0], x=column, bins=50, color='green', label='Target=0')
    sns.histplot(data=merged_data[merged_data['Target'] == 1], x=column, bins=50, color='red', label='Target=1')
    plt.title(t)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.legend(title='Churn', labels=['Остались', 'Ушли'])
    plt.show()

    
#Расчет процентного соотношения  относительно целевой переменной
def percentage(column):
    cross_tab = pd.crosstab(merged_data['Target'],merged_data[column])
    percentages = round(cross_tab.div(cross_tab.sum(axis=0), axis=1) * 100, 1)
    return percentages

# быстрое построение столбчатой диаграммы, отражающее значение категориальной переменной c двумя значениям и доли относительно целевой
def fast_stacked_bar(column_name):
    grouped_data = merged_data.groupby([column_name, 'Target']).size().unstack()

    sns.set(style="darkgrid")
    ax = grouped_data.plot(kind='bar', stacked=True, figsize=(8, 5))

    plt.xlabel(column_name)
    plt.ylabel('Доля')
    plt.title(f'Доля оттока по {column_name}')
    plt.xticks(rotation=0)

    percentages = percentage(column_name)
    for bar in ax.patches:
          ax.text(bar.get_x() + bar.get_width() / 2,
          bar.get_height() / 2 + bar.get_y(),
          round(bar.get_height()), ha = 'center',
          color = 'w', weight = 'regular', size = 10)
    
    plt.legend(title='Churn', labels=['Остались', 'Ушли'])
    plt.show()

    
```

## Стандартизация числовых переменных

Сделаем стандартизацию числовых переменных с помощью StandardScaler для обучающей и тестовой выборки.

```python
numeric = ['MonthlyCharges', 'TotalCharges', 'ContractDuration']
scaler = StandardScaler()
scaler.fit(X_train[numeric])
X_train[numeric] = scaler.transform(X_train[numeric])
X_test[numeric] = scaler.transform(X_test[numeric])
```

## Кодирование категориальных переменных

- Для моделей `Random Forest` и `XGBoost` бинарные категориальные переменные закодируем с помощью OheEncoder, а мультклассовые признаки с помощью OrdinalEncoder
- `CatBoost` сам умеет находить оптимальный режим кодирования, поэтому просто подготовим список признаков для этой модели

```python
binar_columns = ['PaperlessBilling', 'gender', 'Partner', 'Dependents', 'OnlineSecurity', 
               'OnlineBackup', 'DeviceProtection','TechSupport', 'StreamingTV', 
               'StreamingMovies', 'MultipleLines']

multiclass_columns = ['Type', 'PaymentMethod', 'InternetService']
ordinal_columns = ['Type', 'PaymentMethod', 'InternetService']
# В эти списки не включен признак SeniorCitizen, тк он уже закодирован 


# категориальные переменные для модели CatBoost
cat_columns = ['Type', 'PaperlessBilling', 'PaymentMethod', 'gender', 'SeniorCitizen', 
               'Partner', 'Dependents','InternetService', 'OnlineSecurity', 'OnlineBackup',
               'DeviceProtection','TechSupport', 'StreamingTV', 'StreamingMovies', 'MultipleLines']



```

```python
ohe_transformer = OneHotEncoder(drop='first')
ordinal_transformer = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)



encoder = ColumnTransformer(
    transformers=[
        ('ohe', ohe_transformer, binar_columns),
        ('ord', ordinal_transformer, multiclass_columns)],
    remainder='passthrough'  # укажем не преобразовывать другие столбцы, чтобы не потерять SeniorCitizen
)


X_train_encoded = encoder.fit_transform(X_train)
X_test_encoded = encoder.transform(X_test)
```

## Обучение моделей

Напишем функцию для проверки модели на кросс-валидации. Для удобства анализа сохраним результаты тестирования моделей в таблицу.

```python
results_df = pd.DataFrame(columns=['Model', 'ROC-AUC', 'Accuracy'])
```

```python
def scoring(model, features, target, model_name):
    global results_df
    
    kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
    auc_roc_scores = cross_val_score(model, features, target, cv=kf, scoring='roc_auc')
    accuracy_scores = cross_val_score(model, features, target, cv=kf, scoring='accuracy')
    roc_auc_mean = round(auc_roc_scores.mean(), 6)
    accuracy_mean = round(accuracy_scores.mean(), 6)

    results_df = pd.concat([results_df, pd.DataFrame({'Model': [model_name], 'ROC-AUC': [roc_auc_mean], 'Accuracy': [accuracy_mean]})], ignore_index=True)

    
    print("AUC-ROC:", roc_auc_mean)
    print("Accuracy:", accuracy_mean)
```

Обучим и оптимизируем гиперпараметры для 3 моделей: Random Forest, XGBoost и CatBoost. 
Мы выбрали эти модели, так как они широко применяются в задачах классификации и имеют хорошую производительность в работе с табличными данными. 
- Каждую модель обучим сначала «из коробки» — без оптимизации гиперпараметров
- Затем переберем гиперпараметры с помощью Optuna и проверим метрики на кросс-валидации
- Выберем лучшую модель и проверим ее на тестовой выборке

### Random Forest

Обучим модель без оптимизации гиперпараметров.

```python
model_rf = RandomForestClassifier(random_state=RANDOM_STATE)
scoring(model_rf, X_train_encoded, y_train, "Random Forest")
```

Модель показывает скромные результаты метрик, поэтому мы решили улучшить ее производительность с помощью оптимизации гиперпараметров. Для этого мы используем оптимизатор Optuna.

```python
def objective(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 200, step=50),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 10),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 5),
    }
    
    model_rf = RandomForestClassifier(random_state=RANDOM_STATE, **params)
    auc_roc_scores = cross_val_score(model_rf, X_train_encoded, y_train, cv=5, scoring='roc_auc')
    return auc_roc_scores.mean()

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=100)

best_params = study.best_params
print("Best params:", best_params)
```

```python
model_rf_opt = RandomForestClassifier(random_state=RANDOM_STATE, **best_params)
scoring(model_rf_opt, X_train_encoded, y_train, "Random Forest optimized")
```

### CatBoost

Обучим модель `CatBoost` без оптимизации гиперпараметров.

```python
model_cbc = CatBoostClassifier(random_state=RANDOM_STATE, loss_function='Logloss', learning_rate=0.5, verbose=0, cat_features=cat_columns)
scoring(model_cbc, X_train, y_train, 'Catboost')
```

Модель без оптимизации показывает хорошие результаты, попробуем их еще улучшить. Оптимизируем гиперпараметры модели.

```python
def objective(trial):
    params = {
        'depth': trial.suggest_int('depth', 1, 10),
        'l2_leaf_reg': trial.suggest_int('l2_leaf_reg', 0, 10),
        'iterations': trial.suggest_int('iterations', 50, 200, 25),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.5)}

    model_cbc = CatBoostClassifier(random_state=RANDOM_STATE,
                                   loss_function='Logloss',
                                   verbose=0,
                                   cat_features=cat_columns,
                                   **params)

    # Выполняем внутреннюю кросс-валидацию и получаем оценку производительности
    scores_auc = cross_val_score(model_cbc, X_train, y_train, cv=5, scoring='roc_auc')

    # Возвращаем оценку AUC-ROC для оптимизации
    return scores_auc.mean()



study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=150)
```
